```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt 

```

```python
# Create a qunatum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1,1)

# Apply hadmard gate to create superposition
qc.h(0)

# Measure the qubit
qc.measure(0,0)

# Use qiskit's Aer simulator
simulator = Aer.get_backend('aer_simulator')

# Run the curcuit on the simulator
job = simulator.run(qc, shots=1000) # 1000 measurments
result=job.result()

# get the measurment results
counts = result.get_counts(qc)
print(counts)

# convert dictionary keys and values into lists
labels = list(counts.keys()) # ['0','1']
values = list(counts.values()) # [490,510]

# Define colors for bars
colors = ['blue','red']

# plot the bar chart
plt.bar(labels,values,color =colors)

# Add labels and title
plt.xlabel("Measurment outcome")
plt.ylabel("Count")
plt.title("Quantum Measurment Results")
```
